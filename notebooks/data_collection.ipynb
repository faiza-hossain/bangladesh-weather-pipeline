```python
import requests
import pandas as pd
import time
```

```python
#defining target cities & their co-ordinates
bangladesh_major_cities={
    'Dhaka':{
        'lat':23.7104,
        'lon':90.4074
    },
    'Chittagong':{
        'lat':22.3384,
        'lon':91.8317
    },
    'Sylhet':{
        'lat':24.899,
        'lon':91.872
    },
    'Rajshahi':{
        'lat':24.374,
        'lon':88.6011
    }
}
all_city_data=[]
#looping through each city
for city_name,coord in bangladesh_major_cities.items():
  print(f'Fetching weather data for {city_name}.....')
  url=f"https://archive-api.open-meteo.com/v1/archive?latitude={coord['lat']}&longitude={coord['lon']}&start_date=2015-12-31&end_date=2026-01-01&hourly=temperature_2m,relative_humidity_2m,rain,pressure_msl,cloud_cover,wind_speed_100m,wind_direction_100m,soil_temperature_100_to_255cm&models=era5&timezone=auto"
  response=requests.get(url)#hitting the api
  data=response.json()#converting raw web data
  hourly_dict=data['hourly']#extracting the 'hourly' dictionary
  df_city=pd.DataFrame(hourly_dict)
  df_city['city']=city_name
  all_city_data.append(df_city)
  print(f"\nSuccessfully processed {city_name}!")
  time.sleep(1)

print("\nAll cities fetched successfully!")
```

```python
#stacking the dataframes
df_all=pd.concat(all_city_data,ignore_index=True)
df_all.to_csv("bangladesh_weather_2016_2025.csv",index=False)
print("File Saved Successfully!")
```
